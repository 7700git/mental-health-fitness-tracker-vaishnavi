PERFORMING DATA SPLIT OPERATION

1.   training_data : 80%
2.   testing_data : 20%

```python
x = data.drop('mental_fitness', axis=1)
y = data['mental_fitness']

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.20, random_state=2)
print("xtrain: ", xtrain.shape)
print("xtest: ", xtest.shape)
print("ytrain: ", ytrain.shape)
print("ytest: ", ytest.shape)
```

MODEL TRAINING:
We are done with the preprocessing part(EDA,cleaning,visualization for better understanding of our dataset)

we will now be using different training algorithms to train our model and we will test and compare each algorithm's accuracy
1. linear regression
2. Random Forest(Ensemble method)
3. SVM
4. Lasso Regression


LINEAR REGRESSION METHOD

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

lr = LinearRegression()
lr.fit(xtrain, ytrain)

ytrain_pred = lr.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)

rmse = np.sqrt(mean_squared_error(ytrain, ytrain_pred))

r2 = r2_score(ytrain, ytrain_pred)

print("The linear regression model performance for the training set:")
print("_____________________________________________________________")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))
```

Random forest(can be used for both classification ans regression problems)

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain, ytrain)

ytrain_pred = rf.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
r2 = r2_score(ytrain, ytrain_pred)

print("The Random Forest regression model performance for the training set:")
print("_____________________________________________________________")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))
```

SVM(Support vector machine)

1.   SVR(Support Vector Regression)
2.   SVC(Support Vector Classification)

Here we will be employing SVR




```python
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score

svm_reg = SVR(kernel='linear')
svm_reg.fit(xtrain, ytrain)

y_pred = svm_reg.predict(xtrain)
mse = mean_squared_error(ytrain, y_pred)
rmse = (np.sqrt(mean_squared_error(ytrain, y_pred)))
r2 = r2_score(ytrain, y_pred)

print("The SVM regression model performance for the training set:")
print("_____________________________________________________________")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))
```

##lasso Regression-
it is based on a penalty model where absolute value of coeffitients is employed for feature selection.
The regularization parameter must be set, in this case parameter alpha at 0.001 gave the best performance.


```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
lasso_regressor = Lasso(alpha=0.0001)  # You can adjust the regularization parameter (alpha) based on your data
lasso_regressor.fit(xtrain, ytrain)
y_pred = lasso_regressor.predict(xtrain)
mse = mean_squared_error(ytrain, y_pred)
rmse = (np.sqrt(mean_squared_error(ytrain, y_pred)))
r2 = r2_score(ytrain, y_pred)

print("The lasso regression model performance for the training set:")
print("_____________________________________________________________")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))
```
